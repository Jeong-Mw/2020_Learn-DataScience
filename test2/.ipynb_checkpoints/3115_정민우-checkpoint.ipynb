# 주택 가격 예측하기 (회귀)

## 조건 : 테스트세트 점수가 0.7 이상이어야 함 (아래 3가지 알고리즘 중 2가지 선택해서 사용)

### 힌트 : 과소적합이 나올 경우, 특성 늘린 뒤, MinMaxScaler 사용하기

```python
import pandas as pd
import numpy as np
```

## 'housing_price' csv파일을 불러와 'housing' 변수에 저장하기

```python
housing = pd.read_csv('housing_price.csv')
housing.head()
```

## housing 결측치 확인하기

```python
housing.isna().sum()
```

## housing에 결측치가 있을 경우 null이 있는 행 모두 제거하기 + 잘 제거되었는지 확인

```python
# housing.dropna(inplace=True) # 
housing = housing.dropna(inplace=False)
housing.isna().sum()
```

## 문제데이터(X), 정답데이터(y) 나누기
- 정답 데이터(y)는 'median_house_value'
- 문제 데이터(X)는 'median_house_value'을 뺀 나머지

```python
X = housing.drop(['median_house_value'], axis =1, inplace=False)
y = housing['median_house_value']
```

## 문제데이터(X), 정답데이터(y) 잘 나누어졌는지 확인

```python
X.shape, y.shape
```

## EDA
- seaborn의 pairplot 사용
- x축은 total_rooms 부터 median_income까지 나타내기(총 5개 칼럼)
- y축은 median_house_value 
- kind = 'reg' 

```python
import seaborn as sns
```

```python
sns.pairplot(housing, x_vars=housing.columns[3:8], y_vars='median_house_value' , kind='reg', height=7, aspect=0.7)
```

## 학습용/테스트용 데이터 나누기
- test_size = 0.3
- random_state=3

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=3)
```

## 학습용/테스트용 데이터 잘 나누어졌는지 확인

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## 머신러닝 선택+객체만들기+학습+평가(R squared(0.7이상 나와야 함), RMSE 둘 다 계산)

### 힌트 : 과소적합이 나올 경우, 특성 늘린 뒤, MinMaxScaler 사용하기

### 회귀 알고리즘 종류 (3가지 중 2가지 선택)
- LinearRegression
- Ridge
- KNeighborsRegressor

## LinearRegression 알고리즘 사용할 경우

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)


print(f"훈련 점수 : {lr.score(X_train, y_train)}")
print(f"테스트 점수 : {lr.score(X_test, y_test)}")
y_pred = lr.predict(X_test)
print(f"RMSE: {np.sqrt((mean_squared_error(y_test, y_pred)))}")
```

```python
col = X.columns
for i in range(col.size):
    for j in range(i,col.size):
        X[col[i]+'x'+col[j]] = X[col[i]]*X[col[j]]
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
lr2 = LinearRegression()
```

```python
lr2.fit(X_train, y_train)
```

```python
print(f"훈련 점수 : {lr2.score(X_train, y_train)}")
print(f"테스트 점수 : {lr2.score(X_test, y_test)}")
y_pred = lr2.predict(X_test)
print(f"RMSE: {np.sqrt((mean_squared_error(y_test, y_pred)))}")
```

## Ridge 알고리즘 사용할 경우

```python
from sklearn.linear_model import Ridge
ridgeReg = Ridge(alpha=0.00005)
ridgeReg.fit(X_train, y_train)
```

```python
print(f"훈련 점수 : {ridgeReg.score(X_train, y_train)}") 
print(f"테스트 점수 : {ridgeReg.score(X_test, y_test)}")
y_pred= ridgeReg.predict(X_test)
print(f"RMSE: {np.sqrt((mean_squared_error(y_test, y_pred)))}")
```

## KNeighborsRegressor 알고리즘 사용할 경우

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model = KNeighborsRegressor(n_neighbors=11)
model.fit(X_train, y_train)

print("훈련 세트 점수:", model.score(X_train, y_train))
print("테스트 세트 점수:", model.score(X_test, y_test))
y_pred= model.predict(X_test)
print(f"RMSE: {np.sqrt((mean_squared_error(y_test, y_pred)))}")
```
