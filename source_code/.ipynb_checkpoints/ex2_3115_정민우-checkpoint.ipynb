```python
import pandas as pd
```

```python
CCTV_Seoul = pd.read_csv('../data/01. CCTV_in_Seoul.csv', encoding='utf-8')
CCTV_Seoul.head()
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0]:'구별'}, inplace=True)
CCTV_Seoul.head()
```

```python
CCTV_Seoul['최근증가율']=(CCTV_Seoul['2016년']+CCTV_Seoul['2015년']+CCTV_Seoul['2014년'])/CCTV_Seoul['2013년도 이전']*100
CCTV_Seoul.sort_values(by="최근증가율",ascending=False).head()
```

```python
pop_Seoul = pd.read_excel('../data/01. population_in_Seoul.xls',header=2, encoding='utf-8')
pop_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('../data/01. population_in_Seoul.xls',header=2, usecols='B,D,G,J,N',encoding='utf-8')
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={pop_Seoul.columns[0]:'구별', pop_Seoul.columns[1]:'인구수', pop_Seoul.columns[2]:'한국인', pop_Seoul.columns[3]:'외국인', pop_Seoul.columns[4]:'고령자'}, inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul.drop([0],inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul.columns
```

```python
pop_Seoul['구별'].unique()
```

```python
pop_Seoul[pop_Seoul['구별'].isnull()]
```

```python
pop_Seoul.drop([26],inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul['외국인 비율']=pop_Seoul['외국인']/pop_Seoul['인구수']*100
pop_Seoul['고령자 비율']=pop_Seoul['고령자']/pop_Seoul['인구수']*100
pop_Seoul.head()
```

```python
pop_Seoul['외국인 비율']=pop_Seoul['외국인']/pop_Seoul['인구수']*100
pop_Seoul['고령자 비율']=pop_Seoul['고령자']/pop_Seoul['인구수']*100
pop_Seoul.head()
```

```python
pop_Seoul.sort_values(by="고령자 비율",ascending=False).head()
```

```python
CCTV_Seoul.head()
```

```python
pop_Seoul.head()
```

```python
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on='구별')
data_result.head()
```

```python
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
```

```python
data_result.head()
```

```python
data_result.set_index('구별',inplace=True)
data_result.head()
```

```python
corr = data_result.corr(method = 'pearson')
corr
```

```python
import statsmodels.api as sm
result_model = sm.OLS.from_fromula("소계 ~ 인구수", data_result).fit()
result_model.summary()
```
