```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 이미지 업로드

```python
import os
```

```python
train_cats_dir = "../data2/cats_and_dogs_filtered/train/cats/"
train_dogs_dir = "../data2/cats_and_dogs_filtered/train/dogs/"
test_cats_dir = "../data2/cats_and_dogs_filtered/test/cats/"
test_dogs_dir = "../data2/cats_and_dogs_filtered/test/dogs/"
```

```python
os.listdir(train_cats_dir)
```

```python
train_cat_fnames = os.listdir(train_cats_dir)
print("데이터 확인 : ", train_cat_fnames[:3])
print("데이터 개수 : ", len(train_cat_fnames), "\n")

train_dog_fnames = os.listdir(train_dogs_dir)
print("데이터 확인 : ", train_dog_fnames[:3])
print("데이터 개수 : ", len(train_dog_fnames), "\n")

test_cat_fnames = os.listdir(test_cats_dir)
print("데이터 확인 : ", test_cat_fnames[:3])
print("데이터 개수 : ", len(test_cat_fnames), "\n")

test_dog_fnames = os.listdir(test_dogs_dir)
print("데이터 확인 : ", test_dog_fnames[:3])
print("데이터 개수 : ", len(test_dog_fnames), "\n")
```

```python
plt.imread("../data2/cats_and_dogs_filtered/train/cats/cat.0.jpg")
```

```python
def load_images(defaul_path, filenames):
    images = []
    
    for filename in filenames:
        filepath = defaul_path + filename
        images.append(plt.imread(filepath))
        
    return np.array(images)
```

```python
X_train_cats = load_images(train_cats_dir, train_cat_fnames)
X_train_cats.shape
```

```python
X_train_dogs = load_images(train_dogs_dir, train_dog_fnames)
X_train_dogs.shape
```

```python
X_test_cats = load_images(test_cats_dir, test_cat_fnames)
X_test_cats.shape
```

```python
X_test_dogs = load_images(test_dogs_dir, test_dog_fnames)
X_test_dogs.shape
```

## 이미지 확인

```python
figure, axes = plt.subplots(nrows=1, ncols=5)
figure.set_size_inches(18, 5)

axes[0].imshow(X_train_cats[0])
axes[1].imshow(X_train_cats[1])
axes[2].imshow(X_train_cats[2])
axes[3].imshow(X_train_cats[3])
axes[4].imshow(X_train_cats[4])
```

```python
figure, axes = plt.subplots(nrows=1, ncols=5)
figure.set_size_inches(18, 5)

axes[0].imshow(X_train_dogs[0])
axes[1].imshow(X_train_dogs[1])
axes[2].imshow(X_train_dogs[2])
axes[3].imshow(X_train_dogs[3])
axes[4].imshow(X_train_dogs[4])
```

```python
from PIL import Image
```

```python
np.array(Image.fromarray(X_train_cats[3]).resize((224,224)))
```

```python
def resize_image(original_images, size):
    resize_img = []

    for image in original_images:
        resize = np.array(Image.fromarray(image).resize(size))
        resize_img.append(resize)

    return np.array(resize_img)
```

```python
X_train_cats_224 = resize_image(X_train_cats, (224, 224))
print(X_train_cats_224.shape)
X_train_dogs_224 = resize_image(X_train_dogs, (224, 224))
print(X_train_dogs_224.shape)
```

```python
X_test_cats_224 = resize_image(X_test_cats, (224, 224))
print(X_test_cats_224.shape)
X_test_dogs_224 = resize_image(X_test_dogs, (224, 224))
print(X_test_dogs_224.shape)
```

```python
figure, axes = plt.subplots(nrows=1, ncols=5)
figure.set_size_inches(18, 5)

axes[0].imshow(X_train_dogs_224[0])
axes[1].imshow(X_train_dogs_224[1])
axes[2].imshow(X_train_dogs_224[2])
axes[3].imshow(X_train_dogs_224[3])
axes[4].imshow(X_train_dogs_224[4])
```

## 고양이, 강아지 합치기

```python
X_train = np.concatenate((X_train_cats_224, X_train_dogs_224))
X_train.shape
```

```python
X_test = np.concatenate((X_test_cats_224, X_test_dogs_224))
X_test.shape
```

## label 생성(고양이 : 0, 강아지 : 1)

```python
y_train = np.array([0] * 1000 + [1] * 1000)
print(y_train.shape)
print(y_train)
```

```python
y_test = np.array([0] * 500 + [1] * 500)
print(y_test.shape)
print(y_test)
```

# CNN 모델 생성

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
```

```python
model = Sequential()

model.add(Conv2D(filters=128, # 필터의 개수
                 kernel_size=(3, 3), # 필터의 크기
                 activation='relu', # 활성화 함수
                 padding='same', #입력사이즈와 출력사이즈가 같도록 설정
                 input_shape=(224, 224, 3)))

model.add(Conv2D(filters=128, 
                 kernel_size=(3, 3), 
                 activation='relu',
                 padding='same'))

model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

model.add(Conv2D(filters=256, kernel_size=(3, 3), activation='relu',
                 padding='same'))

model.add(Conv2D(filters=256, kernel_size=(3, 3), activation='relu',
                 padding='same'))

model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

model.add(Flatten())

model.add(Dense(units=128, activation='sigmoid'))

model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=5) # validation => 검증데이터
```
